## Running SWMM through Python
Example for SWMM tutorial 1 (see SWMM user manual).
***
**Functions:**
<br>- Writes SWMM project.inp files using a template (template.inp) - searches for placeholder strings in the template and replaces them with data from a pandas dataframe. Dataframes can be created manually, or loaded from .csv files.
<br>- Run SWMM using the specified project.inp file
<br>- Import SWMM results using swmmtoolbox package
<br>- Plot outflows 

***
**Files needed:**
<br>- template.inp
<br>- swmm5.exe
<br>- swmm5.dll
<br>- nodes.csv
<br>- links.csv

**Packages needed:**
<br>- pandas
<br>- numpy
<br>- subprocess
<br>- swmmtoolbox
<br>- swmmpy

**Notes:**
<br>- template.inp files must not have any commas in them (otherwise the importer thinks they are column breaks)
<br>- placeholders in template.inp files must not have any other characters (even spaces) in the same line

### Setup

```python
#Import packages:
import pandas as pd
import numpy as np
import subprocess as sb
import importlib
import swmmpy as sp
from swmmtoolbox import swmmtoolbox as st
import matplotlib.pyplot as plt
#Embed matplotlib figures into notebooks:
%matplotlib inline
```

### Option 1: run SWMM using a pre-existing input file

```python
#Run SWMM with original tutorial .inp file and generate .rpt and .out files   
input_file    = 'tutorial1_original.inp'    #filename of original .inp file
report_file   = 'tutorial1_original.rpt'    #filename to write .rpt file to
output_file   = 'tutorial1_original.out'    #filename to write .out file to

sp.run(input_file, report_file, output_file) #run SWMM as a subprocess
```

```python
#Get & plot run data:
Qoriginal = sp.getQ(output_file)  #load node flow data from output file
Qoriginal.plot()                  #plot
plt.ylabel('node inflow (L/s)')
```
